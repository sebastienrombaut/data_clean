### Clean le site_name

```python
def clean_site_name(site_name):
    result = re.search("^.+\((\d{4}[a-zA-Z]{2})\)", site_name)
    
    try:
        result = result.group(1)
    except:
        0
    return result

df["SITE_NAME_CLEANED"] = [clean_site_name(site_name) for site_name in df["SITE_NAME"]]
df
```

### Supprimer la colonne INVOICE_NAME

```python
df = df.drop('INVOICE_NAME', 1)
```

```python
df
```

```python
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
df
```

### Nombre de jours / facture

```python
def calculate_number_of_day_difference(index):
    return pd.to_datetime(df.END[index], format='%d/%m/%Y', errors='coerce') - pd.to_datetime(df.BEGIN[index], format='%d/%m/%Y', errors='coerce')

max_range = len(df["BEGIN"])

df["NUMBER_OF_DAY_INVOICE"] = [calculate_number_of_day_difference(index) for index in range(max_range)]
df
```

### Consommation quotidienne moyenne

```python
df["PERIOD_EUR"] = df["PERIOD_EUR"].str.replace(",", ".")
df["PERIOD_EUR"] = df["PERIOD_EUR"].str.replace(" ", "")

df["DAILY_AVG_CONSUMPTION"] = df["PERIOD_EUR"].astype(float) / df["NUMBER_OF_DAY_INVOICE"].dt.days
df
```

### Merci d'ajouter également une colonne qui donne, pour chaque antenne, le nombre de factures dont on dispose au total.

```python
def calculate_nb_invoice_per_site_name(site_name):
    return len(df.loc[df['SITE_NAME_CLEANED'] == site_name])

df["NUMBER_INVOICE"] = df["SITE_NAME_CLEANED"].apply(calculate_nb_invoice_per_site_name)
df

```
